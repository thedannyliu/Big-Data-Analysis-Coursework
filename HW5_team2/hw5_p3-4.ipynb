# Problem 1

```python
import category_encoders as ce
def onehot(dataframe,cat_col):
    for col in cat_col:
        encoder = ce.OneHotEncoder(cols= [col])
        dfbin = encoder.fit_transform(dataframe[col])
        dataframe = pd.concat([dataframe,dfbin],axis = 1)
    dataframe = dataframe.drop(columns = cat_col)
    return dataframe
```

```python
import warnings
warnings.filterwarnings("ignore")

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score,f1_score
scaler = StandardScaler()
kf = StratifiedKFold(n_splits=5,random_state=0,shuffle=True)
def problem1(clf,dataframe,judge,name,cat_col,con_col):
    dataframe = onehot(dataframe,cat_col)
    X = dataframe.drop(['y'],axis=1)
    y = dataframe['y']
    f1_list = []
    auc_list = []
    for train_index, valid_index in kf.split(X,y):
        train_x,valid_x = X.iloc[train_index],X.iloc[valid_index]
        train_y, valid_y = y.iloc[train_index],y.iloc[valid_index]
        if (judge ==1 and len(con_col) > 0):
            
            for col  in con_col:
                scaler.fit(train_x[[col]].values)
                train_x[['scale' + str(col)]] = scaler.transform(train_x[[col]].values)
                valid_x[['scale' + str(col)]] = scaler.transform(valid_x[[col]].values)
            train_x = train_x.drop(columns = con_col)
            valid_x = valid_x.drop(columns = con_col)
            
        clf.fit(train_x,train_y)
        predictions = clf.predict(valid_x)
        f1 = f1_score(valid_y, predictions)
        auc_score = roc_auc_score(valid_y, clf.predict_proba(valid_x)[:, 1])
        f1_list.append(f1)
        auc_list.append(auc_score)
    #print(name + " auc score: ",round(sum(auc_list)/5,4))
    print(name + " f1 score: ",round(sum(f1_list)/5,4))
```

```python
import lightgbm as lgb
whole_df_list = [income,arrhy,bank,htru,blastchar,shopper,bio,shrutime,spam,default]
name_list = ['income','arrhy','bank','htru','blastchar','shopper','bio','shrutime','spam','default']
whole_cat_list = [income_cat_col,arrhy_cat_col,bank_cat_col,htru_cat_col,blast_cat_col,shopper_cat_col,bio_cat_col,shrutime_cat_col,spam_cat_col,default_cat_col]
whole_col_list = [income_con_col,arrhy_con_col,bank_con_col,htru_con_col,blast_con_col,shopper_con_col,bio_con_col,shrutime_con_col,spam_con_col,default_con_col]
for i in range(len(whole_df_list)):
    problem1(lgb.LGBMClassifier(application = 'binary'),whole_df_list[i],0,name_list[i],whole_cat_list[i],whole_col_list[i])
```

```python
kf = StratifiedKFold(n_splits=5,random_state=0,shuffle=True)
def problem3_5bin(clf,dataframe,judge,name,cat_col,con_col):
    dataframe = onehot(dataframe,cat_col)
    X = dataframe.drop(['y'],axis=1)
    y = dataframe['y']
    f1_list = []
    auc_list = []
    for train_index, valid_index in kf.split(X,y):
        train_x,valid_x = X.iloc[train_index],X.iloc[valid_index]
        train_y,valid_y = y.iloc[train_index],y.iloc[valid_index]
        if (judge ==1 and len(con_col) > 0):
            for col  in con_col:
                mm_range = train_x[col].max() - train_x[col].min()
                min_value = int(np.floor(train_x[col].min()))
                max_value = int(np.ceil(train_x[col].max()))
                # bin=5
                inter_width = int(np.round(mm_range/5))
                if inter_width != 0:
                  intervals = [i for i in range(min_value, max_value+inter_width, inter_width)]
                  train_x[['binning' + str(col)]] = pd.DataFrame(pd.cut(x=train_x[col],bins=intervals,include_lowest=True,labels=list(range(1,len(intervals)))))
                  valid_x[['binning' + str(col)]] = pd.DataFrame(pd.cut(x=valid_x[col],bins=intervals,include_lowest=True,labels=list(range(1,len(intervals)))))
            train_x = train_x.drop(columns = con_col)
            valid_x = valid_x.drop(columns = con_col)
            
        clf.fit(train_x,train_y)
        predictions = clf.predict(valid_x)
        f1 = f1_score(valid_y, predictions)
        auc_score = roc_auc_score(valid_y, clf.predict_proba(valid_x)[:, 1])
        f1_list.append(f1)
        auc_list.append(auc_score)
    #print(name + " auc score: ",round(sum(auc_list)/5,4))
    print(name + " f1 score: ",round(sum(f1_list)/5,4))
```

```python
import category_encoders as ce

def onehot(dataframe,cat_col):
    for col in cat_col:
        encoder = ce.OneHotEncoder(cols= [col])
        dfbin = encoder.fit_transform(dataframe[col])
        dataframe = pd.concat([dataframe,dfbin],axis = 1)
    dataframe = dataframe.drop(columns = cat_col)
    return dataframe

# def label_encode(dataframe,cat_col):
#     for col in cat_col:
#         encoder = ce.OrdinalEncoder(cols= [col])
#         dfbin = encoder.fit_transform(dataframe[col])
#         dataframe = pd.concat([dataframe,dfbin],axis = 1)
#         print(dfbin)
#     dataframe = dataframe.drop(columns = cat_col)
#     return dataframe

def label_encode(dataframe,cat_col):
    for col in cat_col:
        encoder = ce.OrdinalEncoder(cols= [col])
        dataframe[col] = encoder.fit_transform(dataframe[col])
    return dataframe

def frequency_encode(dataframe,cat_col):
    for col in cat_col:
        encoder = ce.CountEncoder(cols= [col])
        dataframe[col] = encoder.fit_transform(dataframe[col])
    return dataframe

def target_encode(dataframe,cat_col,y):
    for col in cat_col:
        encoder = ce.TargetEncoder(cols= [col])
        dataframe[col] = encoder.fit_transform(dataframe[col],y)
    return dataframe

def betatarget_encode(dataframe,cat_col,y):
    for col in cat_col:
        encoder = ce.TargetEncoder(smoothing=10, cols= [col])
        dataframe[col] = encoder.fit_transform(dataframe[col],y)
    return dataframe

def leaveoneout_encode(dataframe,cat_col,y):
    for col in cat_col:
        encoder = ce.LeaveOneOutEncoder(cols= [col])
        dataframe[col] = encoder.fit_transform(dataframe[col],y)
    return dataframe
```

```python
import warnings
warnings.filterwarnings("ignore")

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score,f1_score
scaler = StandardScaler()
cat_encode_list = [target_encode,betatarget_encode,leaveoneout_encode]
kf = StratifiedKFold(n_splits=5,random_state=0,shuffle=True)

def problem4(clf,dataframe,judge,name,cat_col,con_col):
    dataframe_use = dataframe.copy()
    for cat_encoder in cat_encode_list:

      X = dataframe.drop(['y'],axis=1)
      y = dataframe['y']
      f1_list = []
      auc_list = []
      for train_index, valid_index in kf.split(X,y):
          train_x, valid_x = X.iloc[train_index],X.iloc[valid_index]
          train_y, valid_y = y.iloc[train_index],y.iloc[valid_index]
          if (judge ==1 and len(con_col) > 0):
              train_x = cat_encoder(train_x,cat_col,train_y)
              valid_x = cat_encoder(valid_x,cat_col,valid_y)

              for col in con_col:
                  scaler.fit(train_x[[col]].values)
                  train_x[['scale' + str(col)]] = scaler.transform(train_x[[col]].values)
                  valid_x[['scale' + str(col)]] = scaler.transform(valid_x[[col]].values)
              train_x = train_x.drop(columns = con_col)
              valid_x = valid_x.drop(columns = con_col)
              
          clf.fit(train_x,train_y)
          predictions = clf.predict(valid_x)
          f1 = f1_score(valid_y, predictions)
          auc_score = roc_auc_score(valid_y, clf.predict_proba(valid_x)[:, 1])
          f1_list.append(f1)
          auc_list.append(auc_score)
          dataframe = dataframe_use
      #print(name + " auc score: ",round(sum(auc_list)/5,4))
      print(name + " f1 score: ",round(sum(f1_list)/5,4))
```
